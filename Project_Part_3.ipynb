Describe the data:
What are the observations? How many, and how were they collected?

The observations of the data I am using, is the date and location (coordinates) of the features being collected. There are 25,113 observations in this data. The observations were collected by tallying species presence in specific locations. If there were at least ten individuals of a species observed, the species would be denoted with a 1 signifying presence. If there were not ten individuals of a species observed, the species would be denoted a 0, signifying a lack of presence. Data on the climate at each location was obtained through the Finnish Meteorological Institute.

What are the features? How many, and what do they tell us about the observations

In total, there are 1,487 different species of birds, mammals, small rodents, butterflies, moths, understory vascular plants, and phytoplankton. Each of these species is a feature of the data, and they were marked as either 1 (present) or 0 (not present) for each observation. There were also 4 climate features that were collected for each observation: annual precipitation, annual snow count, annual NAO (North Atlantic Oscillation index) as well as annual temp. This is a total of 1,491 features. For each observation, the features collected tell us what species are present at the location that year, as well as the climate of the location at that year.

What other data or knowledge are used?

The data for each species group were split into three regions of finland. North boreal (NB), Middle Boreal (MB), and South Boreal (SB). This gradient of latitude stretched over 1,200 kilometers.

For the two plots of the selected features, I want to create a barplot of the frequency of occurence for the species over the decades. This will show if there has been a change in the occurence of the species in this area over the 4 decade period.

```python
import seaborn as sns

# I obtain the mean for each of the decades
freq_1 = bird_1_MB['Acanthis flammea'].mean()
freq_2 = bird_2_MB['Acanthis flammea'].mean()
freq_3 = bird_3_MB['Acanthis flammea'].mean()
freq_4 = bird_4_MB['Acanthis flammea'].mean()

# Create a new DataFrame with the decades and the corresponding mean
data = {'Decade': [1, 2, 3, 4], 'Acanthis flammea Frequency': [freq_1, freq_2, freq_3, freq_4]}
Acanthis_flammea_MB = pd.DataFrame(data)

# Plot the data in a barplot
sns.barplot(x='Decade', y='Acanthis flammea Frequency', data=Acanthis_flammea_MB)
```

```python
# I obtain the mean for each of the decades
freq_1 = bird_1_MB['Pica pica'].mean()
freq_2 = bird_2_MB['Pica pica'].mean()
freq_3 = bird_3_MB['Pica pica'].mean()
freq_4 = bird_4_MB['Pica pica'].mean()

# Create a new DataFrame with the decades and the corresponding mean
data = {'Decade': [1, 2, 3, 4], 'Pica pica Frequency': [freq_1, freq_2, freq_3, freq_4]}
Pica_pica_MB = pd.DataFrame(data)

# Plot the data in a barplot
sns.barplot(x='Decade', y='Pica pica Frequency', data= Pica_pica_MB)
```

These visualizations of the Pica pica and Acanthis flammea show the changes in their occurence frequency over the four decades. This data relates to the describe function used in section 3.2 because the bars value corresponds to the mean value given by the describe function for each feature and decade.

We can see from the visualizations that the Pica pica remained relativly constant over the four decades, and for the Acanthis flammea, the frequency remained relatively constant for the first three decades, then dropped by about half in the fourth decade.
